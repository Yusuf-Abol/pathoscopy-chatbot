```python
import json
import shutil

# Create backup
shutil.copy('pathoscopy.ipynb', 'pathoscopy.ipynb.backup')

# Load the notebook
with open('pathoscopy.ipynb', 'r') as f:
    nb = json.load(f)

# Remove widgets metadata entirely (safest for GitHub)
if 'metadata' in nb:
    if 'widgets' in nb['metadata']:
        del nb['metadata']['widgets']
    
    # Also clean up any cell-level widget metadata
    if 'cells' in nb:
        for cell in nb['cells']:
            if 'metadata' in cell and 'widgets' in cell['metadata']:
                del cell['metadata']['widgets']

# Save the fixed notebook
with open('pathoscopy.ipynb', 'w') as f:
    json.dump(nb, f, indent=2)

print("Notebook cleaned! Widget metadata removed.")
```
